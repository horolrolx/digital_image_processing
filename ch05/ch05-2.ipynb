```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
from random import randint
```

```python
# 10% noise
noiseRatio = 10
whiteMax = noiseRatio / 2
blackMax = noiseRatio
```

```python
noiseImg = gImg.copy()

for h in range(gImg.shape[0]):
    for w in range(gImg.shape[1]):
        val = randint(0, 100)
        if val < whiteMax:
            noiseImg[h, w] = 255
        elif val < blackMax:
            noiseImg[h, w] = 0  

plt.imshow(noiseImg, cmap='gray')
```

```python
outImg= cv2.medianBlur(noiseImg, 3)
plt.imshow(outImg, cmap='gray')
```
