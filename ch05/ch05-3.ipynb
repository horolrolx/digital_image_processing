```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
outImg = cv2.GaussianBlur(gImg, (5, 5), 0)
plt.imshow(outImg, cmap='gray')
```

```python
outImg = cv2.GaussianBlur(gImg, (7, 7), 1)
plt.imshow(outImg, cmap='gray')
```

```python
outImg = cv2.GaussianBlur(gImg, (7, 7), 3)
plt.imshow(outImg, cmap='gray')
```

```python
img = cv2.imread('../images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
mean_Img = cv2.blur(gImg, (5, 5))
plt.imshow(mean_Img, cmap='gray')
```

```python
edgeImg = cv2.addWeighted(gImg, 1.0, mean_Img, -1.0, 0)
plt.imshow(edgeImg, cmap='gray')
```

```python
outImg = cv2.addWeighted(gImg, 1.0, edgeImg, 2.0, 0)
plt.imshow(outImg, cmap='gray')
```
